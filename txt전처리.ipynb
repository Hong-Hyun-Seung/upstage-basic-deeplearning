```python
import kss
import konlpy
from konlpy.tag import Mecab
mecab = Mecab()
```

### morph_filter 만들어 두기
-> mecab으로 우리 나라말은 문장 안에 명사, 동사, 형용사를 다 포함해야 온전한 문장이므로, 문장이 3 요소를 다 포함하지 않고 있으면 삭제해주는 함수입니다

```python
def morph_filter(texts): #우리 나라말은 문장 안에 명사, 동사, 형용사를 다 포함해야 온전한 문장이므로, 문장이 3 요소를 다 포함하지 않고 있으면 삭제
    """
    명사(NN), 동사(V), 형용사(J)의 포함 여부에 따라 문장 필터링
    """
    NN_TAGS = ["NNG", "NNP", "NNB", "NP"]
    V_TAGS = ["VV", "VA", "VX", "VCP", "VCN", "XSN", "XSA", "XSV"]
    J_TAGS = ["JKS", "J", "JO", "JK", "JKC", "JKG", "JKB", "JKV", "JKQ", "JX", "JC", "JKI", "JKO", "JKM", "ETM"]

    preprocessed_text = []
    for text in texts:
        morphs = mecab.pos(text, join=False)

        nn_flag = False
        v_flag = False
        j_flag = False
        for morph in morphs:
            pos_tags = morph[1].split("+")
            for pos_tag in pos_tags:
                if not nn_flag and pos_tag in NN_TAGS:
                    nn_flag = True
                if not v_flag and pos_tag in V_TAGS:
                    v_flag = True
                if not j_flag and pos_tag in J_TAGS:
                    j_flag = True
            if nn_flag and v_flag and j_flag:
                preprocessed_text.append(text)
                break
    return preprocessed_text
```

### 해당 부분에서는, 1.file을 읽어 오고, 2.readline으로 line을 읽습니다.3.line에 '.'이 있다면, split을 해주고여 -> 암튼 결과적으로 이 처리를 해준게 sentences라는 배열에 들어가게 됩니다

```python
sentences = []
f = open("/content/drive/MyDrive/Colab Notebooks/증강 현실.txt", "r")
while True:
  line = f.readline()
  if not line:
    break
  split_line = line.split(".")
  for sp in split_line:
    sentences.append(sp)
f.close()
```

문장을 확인해 줍니다

```python
for i, s in enumerate(sentences):
  print(i, s)
```

문장 안에 "\n"혼자 오는것들이 있는데, 이것들 삭제 해줍니다(사실morph filter로 없어지는데, 속도면에서 이득이라 해줍니다)

```python
sentences = [s for s in sentences if s != "\n"]
```

처리된 sentences를 확인해 줍니다

```python
sentences
```

split후, 문장을 KSS로 나눠줍니다 -> 해당 결과는 sents안에 담깁니다

```python
sents = []

for sent in sentences:
  sent = sent.strip()
  if sent:
    splited_sent = kss.split_sentences(sent)
    sents.extend(splited_sent)
```

Sents를 확인해줍니다

```python
for i, sent in enumerate(sents):
  print(i, sent)
```

Sents를 morph 필터 안에 넣어 줘서 한국말이 성립하는 문장을 최종적으로 final_sent안에 넣어줍니다

```python
final_sent = morph_filter(sents)
for i, sent in enumerate(final_sent):
    print(i, sent)
```

final_sent를 확인했고, 이를 파일명에 어울리는 변수에 데이터 프레임으로 할당해 줍니다
